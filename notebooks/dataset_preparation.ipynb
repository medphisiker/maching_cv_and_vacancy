## Исследуем датасет вакансий

```python
dataset_path = "../data/raw/hhparser_vacancy.csv"
df = pd.read_csv(dataset_path)
```

```python
df.head(5)
```

```python
text = df["description"].head(5)[0]
text
```

```python
from lxml import etree

def remove_html_tags(text):
    parser = etree.HTMLParser()
    tree = etree.fromstring(text, parser)
    return etree.tostring(tree, encoding='unicode', method='text')
```

```python
remove_html_tags(text)
```

# Почистим весь датасет

## Проверим на Nan значения

```python
df.isna().sum()
```

```python
len(df)
```

## Проверим на пустые строки

```python
empty_string_counts = df.apply(lambda x: (x == '').sum())
print(empty_string_counts)
```

Для нас самыми важными являются колонки:

* `name` - название должности. Основываясь на них мы будем обучать LM для генерации эмбедингов близких к другу для похожих проффесий и далеких для сильно отличающихся.
* `description` - текстовое описание вакансии, именно по нему наша LM будет искать подходящие резюме для вакансий и наоборот. Если данный столбец имеет пропуск, мы ничего не сможем сделать. Поэтому просто выбросим из базы вакансии где нет текстового описания. Таких всего 2.



```python
df = df.dropna(subset=["description"])
```

```python
len(df)
```

```python
df.isna().sum()
```

```python
empty_string_counts = df.apply(lambda x: (x == '').sum())
print(empty_string_counts)
```

Отлично видим что и правда удалили только две вакансии и теперь все вакансии имеют текстовые описания.

## Удалим html-теги из датасета вакансий

```python
df["description"] = df["description"].apply(lambda text: remove_html_tags(text))
```

```python
df
```

## Сохраним подготовленный датасет

```python
dataset_prepared_path = "../data/processed/hhparser_vacancy_prep.csv"
df.to_csv(dataset_prepared_path, index=False)
```

```python
# убедимся, что датасет сохранен корректно
df = pd.read_csv(dataset_prepared_path)
df
```

Датасет сохранился корректно
